```python
# 3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」

r_ = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=false')
response_ = r_.text
data_ = json.loads(response_)
data_
```

```python
for a in data_:
    likeCount=(a['likeCount'])
    print(likeCount)
```

```python
#非熱門文章平均按讚人數
lst=[]
for a in data_:
    likeCount=(a['likeCount'])
    lst.append(likeCount)
s = sum(lst)
ans = s/30
print(ans)
```
