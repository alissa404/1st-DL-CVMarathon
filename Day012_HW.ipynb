### ① 取出今天所有的發文

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html5lib")
soup
```

```python
# 先找到第一篇文章標題的內容
'''
<!--part_list_2 文字列表 開始-->

<div class="part_list_2">

<h3><span class="date">2020/09/28 19:05</span>
<em class="tag c_society">社會</em>
<a href="/news/20200928/1820129.htm" target="_blank">再揪黑心業者！網售醫療口罩...大陸貨補MIT鋼印扣47萬片</a></h3>
'''
```

```python
for d in soup.find(class_="part_list_2").find_all('h3'):
    print(d.find(class_="date").text, d.find_all('em')[-1].text, d.find_all('a')[-1].text)
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
a={}
for d in soup.find(class_="part_list_2").find_all('h3'):
    data = d.find(class_="date").text, d.find_all('em')[-1].text, d.find_all('a')[-1].text
    date = data[0]
    tag = data[1]
    title = data[2]
    
    # dic.setdefault(key,default_value) 等同于 dic[key] = dic.get(key,default_value)
    # setdefault() 和 get()方法類似, 如果key不存在于字典中，将会添加键并将值设为默认值。
    a.setdefault(tag, [])
    a[tag].append({
        'title': title,
        'date': date
    })
print(a)
```

### ③ 哪一個類別的文章最多

```python
data = []

for i in a:
    d = {}
    d['tag'] = i
    d['count'] = len(a[i])
    data.append(d)

for d in sorted(data, key=lambda d: d['count'], reverse=True):
    ans = list(d.values())
    print(ans)
```
